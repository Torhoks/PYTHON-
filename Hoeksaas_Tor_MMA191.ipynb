## TASK 4 : Provide basic descriptives of the dataset and it's content via short text annotations and summary statistics.

```python
# Provide basic descriptives of counts, min, max, mean and quartiles, STD etc it helps us understand the distribution 
# of the data for each categorical variable. 
RD.describe()
```

```python
#to get further details of the data type for each variable in our dataset.
# we can see the type for each variable, total entries and columns, here we can see our data consist mainly of floats
RD.info() 
```

```python
# as evidenced in the above description there might be some outliers, if we want to check for them in a visual way 
#a boxplot is a good way of identifiying them. in this example I made a boxplot for one variable "Weekly_sales"
#we can clearly see the points who are far away. 

%matplotlib inline 

sns.set(style="whitegrid")
plt.figure(figsize=(10,8))
ax = sns.boxplot(x='Weekly_Sales', data=RD, orient="v")
```

```python
#Another usefull graphical representation is a Histogram in this case for a specific variable
#combined with the basic statistical descriptions shown abovem it gives a meaningfull representation
#here we see where the data is clustered and the values away from the distribution

plt.figure(figsize=(14,8))
sns.distplot(RD['Weekly_Sales'], kde=False)
```

```python
RD.iloc[:,:].corr() 
# we can use a correlation matrix to see how our categorical variables correlate to each other. 
```

```python
#Let's say we have an american client and it's easier for them to have temperature in Farenheit 
#we can add another column Temperature_F with the farenheid convertion of Temperature. 

RD['Temperature_F'] = (RD['Temperature'] * 9/5) + 32
RD.head()
```

## TASK 6 Perform data slicing.

```python
#Let's suppose we want to extract only the data for the MarkDown's and store it as Markdown_all

markdown_all = RD.iloc[:,5:10]
markdown_all.head()
```

```python
#Let's also assume you also want to only extract data for all columns but only half of the rows arbitrarely and 
#store as half_data
half_data = RD.iloc[:75,:]
half_data.info()
#We can see now we have only 75 entries (about half) and 13 columns (since we added temperature_F from original data)
```

## TASK 7: Perform data masking.

```python
#We perform masking by adding a specific criteria like such as >,<, ==
```

```python
#Let's suppose I want to extract data for when fuel Price is > 3 and analyze the values of specific columns 
#for Specific analysis. 
Fuel_3 = RD.loc[RD.Fuel_Price > 3,['Fuel_Price','Weekly_Sales','CPI','Unemployment']]
Fuel_3.head()

```
